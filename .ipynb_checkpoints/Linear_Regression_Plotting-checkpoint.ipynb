```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from scipy import stats

from IPython.core.display import display, HTML                                    
display(HTML("<style>.container { width:100% !important; }</style>"))                                  #change width of Jupyer Notebook to use the whole window res
```

```python
def main():
    #data arrays
    data_title = "TEMPERATURE REGRESSION - NANO 33 BLE SENSE"
    #X_array = np.array([30.44, 25.23, 18.21, 25.72, 29.68, 27.48, 24, 16.3])              #Temperature nicla sense me
    X_array = np.array([28.14,21.85,15.3,23.4,28.8,25.55,22.9,14.98])                     #Temperature nano 33 ble sense
    Y_array = np.array([25, 22, 12.5, 22, 25, 23, 19, 12])                                #Temperature commercial sensor
      
    #X_array = np.array([33.8,42.8,43.2,46.7,34.31,42.5,40.61,44.64])                      #Humidity nano 33 ble sense   
    #X_array = np.array([23,28,35,34,24,30,29,40])                                          #Humidity nicla sense me
    #Y_array =np.array([32,36,51,41,33,38,41,57])                                           #Humidity commerical sensor
    
    #convert array to pandas dataframe
    df = pd.DataFrame({'Temp Nano 33 Ble Sense': X_array, 'Temp Air Monitor': Y_array})
    print(df, "\n")

    #seaborn scatter plot
    sns.set_theme(color_codes=True)
    scatter = sns.relplot(kind="scatter", x="Temp Nano 33 Ble Sense", y="Temp Air Monitor", hue="Temp Nano 33 Ble Sense", 
                                 style="Temp Air Monitor", dashes=False,  data=df, height = 6, aspect = 1.3, s=150)
    scatter._legend.remove()                                                                                                                  #disable seaborn legend to have plot take care of it with location later
    plt.title(data_title, fontsize=16)                                                                                          #add plot title
    plt.subplots_adjust(top=0.95)                                                                                                             #add extra space at the top of the plot for the title

    #regression part
    slope, intercept, r_value, p_value, std_err = stats.linregress(df['Temp Nano 33 Ble Sense'], df['Temp Air Monitor'])
    line = slope*(df['Temp Nano 33 Ble Sense'])+intercept
    plt.plot(df['Temp Nano 33 Ble Sense'], line, '#F95700FF')
     #add custom annotation/legend for the best fit equation
    #legend
    plt.legend(loc='lower right')                                                                                                               #add data legend in the lower right corner
    y_offset = 0    #needed for proper text annotation placement
    plt.text(min(X_array), max(Y_array)-y_offset, 'Best Fit: y={:.4f}x+{:.4f} \nStandard Error: {:.5f}'.format(slope,intercept,std_err), size=15,  
             horizontalalignment='left', color='#F1F4FFFF', weight='semibold', bbox=dict(boxstyle='round', facecolor='#F95700FF', alpha=0.6))
             

    print("Equation: ", 'y={:.2f}x+{:.2f}'.format(slope,intercept))
    print("Standard Error: ", std_err)
    plt.show()

    choice = int(input("Would you like to save the plot? \nPress 1 to save! Press any key to exit.\n"))
    if choice == 1: 
        name = input("Please type a name for the file: ") 
        scatter.savefig('./plots/' + name,dpi=600) 
        print("The plot has been successfully saved!")
    else: print("You have chosen not to save the plot!")

if __name__ == "__main__":                                                                                                                     #guard boilerplate
    main()
```
